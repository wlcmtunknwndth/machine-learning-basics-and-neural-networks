Сделайте это двумя способами: с помощью метода `arange` из numpy и с помощью преобразования списка (list) в numpy array

```python
def concatenate_matrices(a, b):
    vertical = np.vstack((a, b))
    horizontal = np.hstack((a, b))
    return vertical, horizontal
```

## Задача 2.10

Создайте массив из случайных чисел размера 3 на 4 (3 строки и 4 столбца) и сохраните его в переменную random_array. Найдите среднее, медиану, стандартное отклонение созданной матрицы random_array по строкам и по столбцам. По одной строке на среднее, медиану, и стандартное отклонение для операций по строкам и по столбцам - всего должно получиться 6 строк кода. Результатом каждой из этих операций должен быть вектор. Какова длина вектора в каждом из этих случаев?

```python
def analyze_random_array():
    np.random.seed(42)
    random_array = np.random.rand(3, 4)
    
    mean_rows = np.mean(random_array, axis=1)
    mean_cols = np.mean(random_array, axis=0)
    median_rows = np.median(random_array, axis=1)
    median_cols = np.median(random_array, axis=0)
    std_rows = np.std(random_array, axis=1)
    std_cols = np.std(random_array, axis=0)
    
    return random_array, mean_rows, mean_cols, median_rows, median_cols, std_rows, std_cols
```

```python
random_array, mean_rows, mean_cols, median_rows, median_cols, std_rows, std_cols = analyze_random_array()

print(f"Исходный массив:\n{random_array}\n")
print(f"Среднее по строкам: {mean_rows} (длина: {len(mean_rows)})")
print(f"Среднее по столбцам: {mean_cols} (длина: {len(mean_cols)})")
print(f"Медиана по строкам: {median_rows} (длина: {len(median_rows)})")
print(f"Медиана по столбцам: {median_cols} (длина: {len(median_cols)})")
print(f"Стандартное отклонение по строкам: {std_rows} (длина: {len(std_rows)})")
print(f"Стандартное отклонение по столбцам: {std_cols} (длина: {len(std_cols)})")
```

## Задача 2.11

Еще раз найдите только средние значения по строкам у созданной в задании 2.10 матрицы random_array (одна строка кода). Вычтите полученные средние значения из соответствующих строк исходной матрицы random_array. Проверьте, что после вычитания средних среднее по каждой строке стало равно 0. Результат - вектор.

```python
def center_rows():
    np.random.seed(42)
    random_array = np.random.rand(3, 4)
    
    mean_rows = np.mean(random_array, axis=1, keepdims=True)
    centered_array = random_array - mean_rows
    new_means = np.mean(centered_array, axis=1)
    
    return random_array, centered_array, new_means
```

```python
random_array, centered_array, new_means = center_rows()
    
print(f"Исходный массив:\n{random_array}\n")
print(f"Массив после центрирования:\n{centered_array}\n")
print(f"Новые средние по строкам: {new_means}")

is_zero = np.allclose(new_means, 0)
print(f"\nСредние равны нулю: {is_zero}")
```

## Задача 2.12

Пусть мы имеем 2 массива (заданы ниже). Требуется создать массив, который получает результат поэлементно:

True, если элементы 1-го массива с четными индексами больше соответствующих элементов 2-го массива и

True, если элементы 2-го массива с нечетными индексами больше соответствующих элементов 1-го массива.

В противном случае возвращать False.

Реализовать задачу с помощью numpy

Пример массивов и результата:

a = [7, 3, 5, 9, 2, 5, 4, 8, 9, 0]

b = [2, 8, 6, 7, 1, 3, 2, 9, 5, 1]

res = [ True, True, False, False, True, False, True, True, True, True]

Вы можете сделать задачу с помощью слайсинга или комбинации функций np.arange и np.where

```python
def compare_arrays(a, b):
    a = np.array(a)
    b = np.array(b)
    res = np.zeros(len(a), dtype=bool)
    
    even_indices = np.arange(0, len(a), 2)
    odd_indices = np.arange(1, len(a), 2)
    
    res[even_indices] = a[even_indices] > b[even_indices]
    res[odd_indices] = b[odd_indices] > a[odd_indices]
    
    return res
```

```python
a = [7, 3, 5, 9, 2, 5, 4, 8, 9, 0]
b = [2, 8, 6, 7, 1, 3, 2, 9, 5, 1]
print(compare_arrays(a, b))
```
