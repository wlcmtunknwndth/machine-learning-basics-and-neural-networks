```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import cv2
import sklearn
import PIL
import torch
import torchvision
from matplotlib import cm

import warnings
warnings.filterwarnings('ignore')
```

```python
import torch
import torch.nn.functional as F
import torchvision
import os
from torch import nn
from tqdm import tqdm
```

```python
device = 'cpu'
```

```python
transform = torchvision.transforms.Compose(
    [
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize((0.1307,), (0.3081,))
    ]
)
```

```python
def seed_everything(seed):
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
```

```python
seed_everything(42)

mnist_train = torchvision.datasets.MNIST(
    './mnist/',
    train=True,
    download=True,
    transform=transform
)

mnist_val = torchvision.datasets.MNIST(
    './mnist/',
    train=False,
    download=True,
    transform=transform
)

batch_size = 128
train_dataloader = torch.utils.data.DataLoader(mnist_train, batch_size=batch_size, shuffle=True)
val_dataloader = torch.utils.data.DataLoader(mnist_val, batch_size=batch_size, shuffle=True)
```

```python
plt.imshow(mnist_train[0][0].squeeze(0).numpy().reshape([28, 28]))
```

В результате экспериментов было выяснено, что для достижения необходимого качества достаточно поменять оптимизатор. Но по заданию надо поменять всё, поэтому в качестве loss функции был выбран MultiLabelSoftMarginLoss.

```python
model = nn.Sequential(
    nn.Conv2d(in_channels=1, out_channels=12, kernel_size=5),  # grayscale
    nn.ReLU(),
    nn.MaxPool2d(kernel_size=4),   # сократим изображение в 4 раза по каждой из сторон
    nn.Flatten(),
    nn.Linear(6*6*12, 128),
    nn.ReLU(),
    nn.Linear(128, 10)
).to(device=device)

optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
def calc_accuracy_and_loss(model, dataloader, device):
  loss_list = []
  accuracy_list = []
  with torch.no_grad():
      for x, y_target in tqdm(dataloader):
          x, y_target = x.to(device=device), y_target.to(device=device)
          y_pred = model(x)
          loss = F.multilabel_soft_margin_loss(y_pred, F.one_hot(y_target, num_classes=10))
          loss_list.extend([loss.cpu().item()] * x.size(0))
          accuracy_list.extend((torch.argmax(y_pred, dim=-1) == y_target).cpu().numpy().tolist())
  return np.mean(accuracy_list).item(), np.mean(loss_list).item()
```

```python
def train(model, optimizer, device=device, n_epochs=5):
    train_loss = []
    val_loss = []

    train_acc = []
    val_acc = []

    for epoch in range(n_epochs):

        # обучение
        for x_train, y_train in tqdm(train_dataloader):
            x_train, y_train = x_train.to(device=device), y_train.to(device=device)
            y_pred = model(x_train)
            loss = F.multilabel_soft_margin_loss(y_pred, F.one_hot(y_train, num_classes=10))
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()

        cur_train_acc, cur_train_loss = calc_accuracy_and_loss(model, train_dataloader, device)
        cur_val_acc, cur_val_loss = calc_accuracy_and_loss(model, val_dataloader, device)

        train_loss.append(cur_train_loss)
        train_acc.append(cur_train_acc)
        val_loss.append(cur_val_loss)
        val_acc.append(cur_val_acc)

        # валидация
        if epoch % 2 == 0:
            print(f'Epoch: {epoch}, loss: {cur_val_loss}, accuracy: {cur_val_acc}')

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(train_loss, label='Train')
    plt.plot(val_loss, label='Validation')
    plt.xlabel('Epochs', fontsize=16)
    plt.ylabel('Loss', fontsize=16)
    plt.legend(loc=0, fontsize=16)
    plt.grid('on')
    plt.subplot(1, 2, 2)
    plt.plot(train_acc, label='Train')
    plt.plot(val_acc, label='Validation')
    plt.xlabel('Epochs', fontsize=16)
    plt.ylabel('Accuracy', fontsize=16)
    plt.legend(loc=0, fontsize=16)
    plt.grid('on')
    plt.show()
    return train_loss, val_loss, train_acc, val_acc
```

```python
print('Количество обучаемых параметров:', sum(p.numel() for p in model.parameters()))
```

```python
train(model, optimizer)
```

```python
transform_simple = torchvision.transforms.Compose(
    [
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize((0.1307,), (0.3081,)),
    ]
)

transform = torchvision.transforms.Compose(
    [
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize((0.1307,), (0.3081,)),
        torchvision.transforms.RandomRotation(10)
    ]
)
```

```python
seed_everything(42)

mnist_train = torchvision.datasets.MNIST(
    './mnist/',
    train=True,
    download=True,
    transform=transform
)

mnist_val = torchvision.datasets.MNIST(
    './mnist/',
    train=False,
    download=True,
    transform=transform_simple
)

batch_size = 128
train_dataloader = torch.utils.data.DataLoader(mnist_train, batch_size=batch_size, shuffle=True)
val_dataloader = torch.utils.data.DataLoader(mnist_val, batch_size=batch_size, shuffle=True)
```

```python
model = nn.Sequential(
    nn.Conv2d(in_channels=1, out_channels=12, kernel_size=3),  # grayscale
    nn.ReLU(),
    nn.Conv2d(in_channels=12, out_channels=24, kernel_size=3),
    nn.ReLU(),
    nn.MaxPool2d(kernel_size=4),   # сократим изображение в 4 раза по каждой из сторон
    nn.Flatten(),
    nn.Linear(6*6*24, 128),
    nn.ReLU(),
    nn.Linear(128, 10)
).to(device=device)

optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
print('Количество обучаемых параметров:', sum(p.numel() for p in model.parameters()))
train(model, optimizer)
```

Максимальное значение accuracy стало сильно лучше.
