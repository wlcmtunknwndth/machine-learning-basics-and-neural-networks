## Инструменты и терминология

**Anaconda** — платформа для работы с данными, объединяющая в себе: 
- Инструменты и среды разработки
- Менеджер пакетов/библиотек
- Диспетчер окружений для разделения проектов

**Jupyter Notebook** — веб-интерфейс для разработки и анализа данных. Он представляет собой документ из ячеек, в каждой из которых могут содержаться вычисления, выводы, текстовые пояснения, картинки, визуализации, формулы и т. д.

> Дополнительно: с документацией и настройкой Jupyter Notebook также можно ознакомиться [по ссылке](https://jupyter-notebook.readthedocs.io/en/stable/notebook.html).

### Ключевые термины Jupyter Notebook

**Ядро (kernel)** — вычислительный движок, который выполняет код, содержащийся в документе ноутбука.

**Ячейка (cell)** — контейнер для кода, который будет выполняться, или для сопутствующей информации в разметке.

*Jupyter Notebook* предоставляет удобный способ экспериментировать с кодом на Python, чередуя эксперименты с заметками и документацией. Таким образом, проводить необходимые эксперименты и анализ можно не прибегая к командной строке, а полученный файл можно легко опубликовать и поделиться им с другими людьми. 

*Jupyter Notebook* состоит из нескольких «ячеек», расположенных на странице сверху вниз. Ячейки могут содержать текст или код в них. Вы можете изменить тип ячейки, используя меню Cell в верхней части страницы; перейдите в `Cell > Cell type` и выберите либо `Code` для кода Python, либо `Markdown` для текста. Вы также можете изменить это для текущей ячейки, используя выпадающее меню на панели инструментов.

#### Текстовые ячейки
Создайте новую ячейку, измените ее тип на `Markdown`, введите что-нибудь и нажмите `Ctrl-Enter`. Jupyter Notebook отобразит его на странице в отрисованном формате. Вы можете нажать `Enter` или щелкнуть по ячейке, чтобы снова отредактировать ее содержимое. Текст в ячейках `Markdown` отображается в соответствии с набором соглашений, называемых Markdown. Markdown — это простой язык для разметки текста с помощью базовой информации о форматировании текста (такой как жирный шрифт, курсив, гиперссылки, таблицы и т. д.). [Вот руководство](https://www.markdowntutorial.com/).

### Полезные команды (кратко)

- `Enter` — редактировать текст
- `Esc` — выход из режима редактирования в сommand mode
- `Ctrl-Enter`/`Cmd-Enter` — выполнить ячейку
- `Shift-Enter` — выполнить ячейку и перейти в следующую
- Список буквенных шорткатов (в command mode):
    - `m` — преобразовать кодовую ячейку в ячейку для разметки
    - `b` — вставка кодовой ячейки ниже
    - `a` — вставка кодовой ячейки выше
    - `x` — вырезать ячейку
    - `c` — скопировать ячейку
    - `v` — вставить ячейку ниже
    - `h` — вывести подсказку на экран

### - Command mode (press Esc to enable)

В jupyter есть command mode — командный режим, где можно вводить команды для быстрого редактирования ячеек. Его можно активировать кнопкой Esc, после чего возможны варианты. Esc + h откроет справку.

В ней перечислено много команд, из которых чаще всего используются такие:

- `Esc` — выход из режима редактирования (переход в командный режим)
- `Esc + H` — вывести подсказку на экран

- `Esc + F` — найти и заменить;


- `Esc + I` — прервать исполнение;
- `Esc + 0` — перезапустить ядро;


- `Esc + Shift + ↑↓` — выделить ячейку выше/ниже текущей;


- `Esc + Ctrl + Enter` — запустить выделенные ячейки (в естественном порядке);


- `Esc + D + D` — удалить все выделенные ячейки (включая текущую);
- `Esc + Z` — отменить удаление ячеек;
- `Esc + C` — скопировать выделенные ячейки;
- `Esc + X` — вырезать выделенные ячейки;
- `Esc + V` — вставить ранее скопированные ячейки;
- `Esc + A` — вставить ячейку выше;
- `Esc + B` — вставить ячейку ниже;
- `Esc + Shift + M` — объединить выделенные ячейки (или текущую ячейку с нижней);


- `Esc + M` — заменить код на markdown;
- `Esc + Y` — наоборот, markdown —> код.

### - Edit mode (press Enter to enable)

Edit mode - режим редактирования текста в ячейках.
Эти сокращения доступны в редакторе. Здесь всё стандартно.

- `Enter` — редактировать текст
- `Shift + Tab` — показать описание функции, курсор должен стоять сразу после открывающей скобки;


- `Ctrl + /` — закомментировать строку;
- `Ctrl + D` — удалить строку;


- `Ctrl + ↑` — перейти к началу ячейки;
- `Ctrl + ↓` — перейти к концу ячейки;
- `Ctrl + Shift + -` — разбить ячейку по текущей строке;


- `Shift + ↑↓` — выделить ячейку выше/ниже текущей;
- `Ctrl + Enter` — запустить выделенные ячейки;
- `Shift + Enter` — запустить ячейку и перейти на одну ниже, что удобно, когда последовательно запускаете сразу много.
