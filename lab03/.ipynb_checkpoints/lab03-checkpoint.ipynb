### 1. Какова доля семей, в которых минимальный возраст меньше 20 (семьи с детьми)?

```python
df['LastName'] = df['Name'].str.split(',').str[0]

family_min_age = df.groupby('LastName')['Age'].min().reset_index()
family_min_age.columns = ['LastName', 'MinAge']

families_with_children = family_min_age[family_min_age['MinAge'] < 20]
total_families = len(family_min_age)

proportion = len(families_with_children) / total_families

print(f"Количество всех семей: {total_families}")
print(f"Количество семей с детьми (мин. возраст < 20): {len(families_with_children)}")
print(f"Доля семей с детьми: {proportion:.4f} ({proportion*100:.2f}%)")
```

### 2. Какова доля выживших пассажиров класса 3? А пассажиров класса 1?

```python
# Доля выживших пассажиров 3 класса
class_3_survived = df[(df['Pclass'] == 3) & (df['Survived'] == 1)]
class_3_total = df[df['Pclass'] == 3]
survival_rate_class_3 = len(class_3_survived) / len(class_3_total)

# Доля выживших пассажиров 1 класса
class_1_survived = df[(df['Pclass'] == 1) & (df['Survived'] == 1)]
class_1_total = df[df['Pclass'] == 1]
survival_rate_class_1 = len(class_1_survived) / len(class_1_total)

print("=== ДОЛЯ ВЫЖИВШИХ ПАССАЖИРОВ ===")
print(f"Пассажиры 3 класса:")
print(f"  Всего пассажиров: {len(class_3_total)}")
print(f"  Выжило: {len(class_3_survived)}")
print(f"  Доля выживших: {survival_rate_class_3:.4f} ({survival_rate_class_3*100:.2f}%)")

print(f"\nПассажиры 1 класса:")
print(f"  Всего пассажиров: {len(class_1_total)}")
print(f"  Выжило: {len(class_1_survived)}")
print(f"  Доля выживших: {survival_rate_class_1:.4f} ({survival_rate_class_1*100:.2f}%)")
```

### 3. Сколько пассажиров выжило, а сколько - нет?

```python
survival_counts = df['Survived'].value_counts().sort_index()
total_passengers = len(df)

print("=== СТАТИСТИКА ВЫЖИВАНИЯ НА ТИТАНИКЕ ===")
print(f"Общее количество пассажиров: {total_passengers}")

print(f"\nВыжившие (Survived = 1): {survival_counts.get(1, 0)} пассажиров")
print(f"Погибшие (Survived = 0): {survival_counts.get(0, 0)} пассажиров")

# Расчет долей
survival_rate = survival_counts.get(1, 0) / total_passengers
death_rate = survival_counts.get(0, 0) / total_passengers
```

### Задача 2.1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.

```python
all_students = []

base_path = '/Users/artyem.petrov/dev/university/4-1/machine-learning-basics-and-neural-networks/25-26 ОМОиНС (студентам)/Data'

for i in range(0, 10):
    info_file = f"{base_path}/Students_info_{i}.csv"
    marks_file = f"{base_path}/Students_marks_{i}.csv"
    
    info_df = pd.read_csv(info_file)
    marks_df = pd.read_csv(marks_file)
    
    merged_df = pd.merge(info_df, marks_df, on='index')
    merged_df['group'] = i
    
    all_students.append(merged_df)

df = pd.concat(all_students, ignore_index=True)

df.head()
```

### Задание 2.12. 
**Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
# 1. Для каждой этнической группы - средний балл по чтению (циклом)
start_time = time.time()
ethnic_groups = df['race/ethnicity'].unique()
reading_means_loop = {}

for group in ethnic_groups:
    group_scores = df[df['race/ethnicity'] == group]['reading score']
    reading_means_loop[group] = group_scores.mean()

print("Средний балл по чтению для каждой этнической группы (циклом):")
for group, mean_score in reading_means_loop.items():
    print(f"{group}: {mean_score:.2f}")
loop1_time = time.time() - start_time
print(f"Время выполнения: {loop1_time:.4f} секунд\n")
```

```python
# 2. Для каждого уровня образования - минимальный балл по письму (циклом)
start_time = time.time()
education_levels = df['education'].unique()
writing_mins_loop = {}

for level in education_levels:
    level_scores = df[df['education'] == level]['writing score']
    writing_mins_loop[level] = level_scores.min()

print("Минимальный балл по письму для каждого уровня образования (циклом):")
for level, min_score in writing_mins_loop.items():
    print(f"{level}: {min_score}")
loop2_time = time.time() - start_time
print(f"Время выполнения: {loop2_time:.4f} секунд")
```

### Задание 2.14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
df['Math_PassStatus'] = df['math score'].apply(lambda x: 'P' if x >= 50 else 'F')

math_pass_count = df['Math_PassStatus'].value_counts()

print(f"Результаты экзамена по математике:")
math_pass_count
```

```python
df['Reading_PassStatus'] = df['reading score'].apply(lambda x: 'P' if x >= 50 else 'F')

reading_pass_count = df['Reading_PassStatus'].value_counts()

print("Результаты экзамена по чтению:")
reading_pass_count
```

```python
df['Writing_PassStatus'] = df['writing score'].apply(lambda x: 'P' if x >= 50 else 'F')
writing_pass_count = df['Writing_PassStatus'].value_counts()

print("Результаты экзамена по письму:")
writing_pass_count
```

### Задание 2.15. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
df['OverAll_PassStatus'] = df.apply(
    lambda row: 'P' if (row['math score'] >= 50 and 
                       row['reading score'] >= 50 and 
                       row['writing score'] >= 50) else 'F', 
    axis=1
)

overall_pass_count = df['OverAll_PassStatus'].value_counts()
print("Результаты по всем экзаменам:")

print(overall_pass_count)

print(f"\nКоличество студентов, сдавших все экзамены: {overall_pass_count.get('P', 0)}")
```
